```python
import warnings
warnings.simplefilter("ignore")
import os
from yahoofinancials import YahooFinancials
import yfinance as yf
import numpy as np
import pandas as pd
import time
```

```python
os.chdir('C:\\Users\\rosorio\\Desktop\\ScriptsFin') # sets directory to parent directory 
fh = open('faanmg.txt', 'r')
pth='./sp500/'
os.chdir(pth) #changes directory to sp500

for line in fh:
    
    #reading lines of the text file
    line = line.rstrip('\n')
    firm=line
    print(firm)
    data = yf.download(firm, start="2006-01-01", end="2020-08-29")
    del data['Adj Close']
    data['Name']=firm
    
    #saving to csv file
    data.to_csv("{0}_2006-01-01_to_2020-08-29.csv".format(firm))
    
    time.sleep(1)
    
```
