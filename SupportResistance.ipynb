```python
import numpy as np
import pandas as pd
import yfinance as yf
import trendln
import matplotlib. pyplot as plt
```

```python
tick = yf.Ticker('NVDA') # TICKER
hist = tick.history(period="max", rounding=True)
h = hist[-50:].Close
mins, maxs = trendln.calc_support_resistance(h)
minimaIdxs, pmin, mintrend, minwindows = trendln.calc_support_resistance((hist[-50:].Low, None)) #support only
mins, maxs = trendln.calc_support_resistance((hist[-50:].Low, hist[-50:].High))
(minimaIdxs, pmin, mintrend, minwindows), (maximaIdxs, pmax, maxtrend, maxwindows) = mins, maxs
```

```python
idx = hist[-50:].index
fig = trendln.plot_sup_res_date((hist[-50:].Low, hist[-50:].High), idx) 
plt.savefig('suppres.svg', format='svg')
plt.show()
```
