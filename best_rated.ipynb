```python
import os
import yfinance as yf
import numpy as np
import pandas as pd
import itertools
import datetime
from datetime import date
from datetime import datetime, timedelta
import time
import random
```

```python
column_names = ['firm', 'buyrating', 'number_ratings']
df = pd.DataFrame(columns = column_names)
fh = open('sp500.txt')
i=0

#rating
for firm in fh:
    try:
        firm=firm.rstrip().lstrip()
        firma = yf.Ticker(firm)
        recommend=firma.recommendations
        recomendacion=pd.crosstab(recommend['To Grade'], recommend['To Grade'],margins=True)
        buy       = len(recommend[(recommend['To Grade']=="Buy")])
        strongbuy = len(recommend[(recommend['To Grade']=="Strong Buy")])
        outperform = len(recommend[(recommend['To Grade']=="Outperform")])
        mktoutperform = len(recommend[(recommend['To Grade']=="Market Outperform")])
        overweight = len(recommend[(recommend['To Grade']=="Overweight")])
        positive = len(recommend[(recommend['To Grade']=="Positive")])
        longtermbuy = len(recommend[(recommend['To Grade']=="Long-Term Buy")])
        posopinions = buy + strongbuy + outperform + mktoutperform + overweight + positive + longtermbuy
        buyrating = posopinions / len(recommend)
        df.loc[i, ['firm']]=firm
        df.loc[i, ['buyrating']]=buyrating
        df.loc[i, ['number_ratings']]=len(recommend)
    except:
        df.loc[i, ['buyrating']]='NaN'
    
    i=i+1
    
    x=random.randint(1,2)
    print('Random number waiting...',x, '  Now processing firm ... ', firm)
    time.sleep(x)
    
#closing text file
fh.close()
df=df.sort_values(by=['buyrating'], ascending=False)
df
```
